# 1 A simple plot

```python
from matplotlib import pyplot as plt
```

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 5, 10, 15, 20, 25, 30]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_2, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) #for grids
plt.legend(loc='upper left') #for legends, done at the end, everything can be customised
```

```python
#Things that can be abbreviated
#colour: c
#linestyle: ls
#linewidth: lw
# data
x = [0, 5, 10, 15, 20, 25, 30]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# plots
plt.plot(x, y_2, label='Y values',
         c='b', ls='-', lw = 3) #or even c='b--'for dotted blue line
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) #for grids
plt.legend(loc='upper left')
#plt.show() -- impt to show the graph in python terminal
```

```python
#2 lines in 1 graph
plt.plot(x, y_1, label = 'a')
plt.plot(x, y_2, label = 'b')
plt.legend()
```

```python
plt.plot(x,y_1,
        c = 'red',
        label = 'y-values',
        ls = '-')
plt.legend()
```

## 1.2 Adding another plot

```python
# Not straight lines
x = [0, 2, 4, 6, 8, 10]
y_1 = [1, 2, 4, 9, 17, 22]
y_2 = [0, 3, 5, 8, 12, 15]

#plots
plt.plot(x, y_1, label='Y1 values', color='orange', linestyle='dashed') #might need to be careful abt colour chosen eg. colour blindness; can ref 'colour brewer' for colours that can be used / 'Paul Tol'; some colours grab ppl's attention more sometimes, but colours do not always need to be used, can even differentiate based on lines style
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
x = [0, 5, 10, 15, 20, 25, 30]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green', alpha = 0.5)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
#able to do numpy arrays, able ot do mathematics
# Some data for plotting
x = [0, 5, 10, 15, 20, 25, 30]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1**3, color='red', linestyle='dashed', label='Y values') #able to do math stuff to transform the graph
plt.plot(np_x, np_y_2+1385, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 5, 10, 15, 20, 25, 30])

x2 = x**2              #graph of x**2
sin_x = np.sin(x)      #graph of sin, will not work if you use some other package instead of numpy

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', ) #can you restrict the maximum value of y-axis to eg. 100? plt.ylim([-100, 100])
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

### Using a different axis

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='upper right')    #needed before twinx()  for this graph to have a legend                           

plt.twinx() #new y-axis; 2 graphs in 1-graph, just so happens they share the same x-axis, can actually use twiny for the same y-axis                                         
                                                           
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right') #loc indicates location of where the legends should go
```

```python
x = np.arange(0, 12, .5) #make graphs nicer OR refâ†“
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='upper right')                            

plt.twinx()                                       
                                                           
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')
```

```python
x = np.linspace(0,6,100) 
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc=(.1, .25))       #ways to change legends; anything more than 1 --> out of graph frame                

plt.twinx()                                       
                                                           
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc=(.1, .35)) #such that these are closer to each other
```

# 3 Saving to disc

```python
x = np.linspace(0,6,100) 
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc=(.1, .25))

plt.twinx()                                       
                                                           
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc=(.1, .35))

plt.savefig('plot.png', dpi = 500, transparent = True ) #saving graph #dpi: dense & resolution - have some restrictions sometimes; extensions can be changed to get diff files
#bitmat (eg. png: becomes very pixalated when zoomed) vs vector (eg. pdf: clear no matter how much you zoom)
#transparent = True: background is transparent (but only works for png)
```

## 4.1 Plotting data from files

```python
plt.style.use('default')
filename = 'Complete_TAVG_daily.txt'
data = np.loadtxt(filename, skiprows = 23)
year = data[:,1]
anomaly = data[:,-1]
plt.plot(year, anomaly)
plt.grid(alpha = 0.5)
plt.hlines(0,year[0],year[-1],color = 'red', ls = 'dashed') #draws that straight line at 0
plt.savefig('temperature.png')
```

## 4.2 Styles
